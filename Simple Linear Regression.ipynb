#### Simple exemple en python

Faisons une prédiction sur le salaire d'un développeur basé sur leurs expériences.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X=np.array([0,3,6,8])
Y=np.array([35,45,65,80])
print(X,Y)
```

```python
plt.plot(X,Y,'*')
plt.xlabel("Années d'expérience | notre explicative variable 'x'" )
plt.ylabel("Salaire | Target variable 'y'")
plt.title("Scratter plot / Le complot de l'éparpilleur ")
plt.savefig("./intuitive_scatter.png")
```

```python
def reg_plot(x,y,m):
    plt.scatter(x,y,c='pink',label="nos data")
    # Tracé de la courbe de prédiction en utilisant le modèle 'm' pour prédire les valeurs correspondantes à 'x'
    #x.reshape permet de remodeler un tableau
    # La courbe est tracée en bleu ('blue') et une légende "prediction de la courbe" est ajoutée
    plt.plot(x, m.predict(x.reshape(-1, 1)), color='blue',label="prediction de la courbe")
    plt.xlabel("explicative variable 'x' ")
    plt.ylabel("target variable 'y'")
    plt.legend()
    return None
```

```python
# Importation de la classe LinearRegression depuis la bibliothèque de modèles linéaires de scikit-learn
linear_model = LinearRegression()

# Entraînement du modèle linéaire avec les données X et Y.
# X.reshape(-1, 1) est utilisé pour remodeler le tableau X en une matrice unidimensionnelle
# ayant une seule colonne, ce qui est attendu par la méthode fit.
linear_model.fit(X.reshape(-1, 1), Y)

# Appel à la fonction reg_plot pour tracer un nuage de points et une courbe de prédiction basée sur le modèle linéaire
reg_plot(X, Y, linear_model)

plt.savefig("./approche_intuitive.png")
```

### Générons une ligne de données

```python
import numpy as np
```

```python
x=np.arange(75)
delta = np.random.uniform(-10,10, size=(75,))
y = 0.4 * x +3 + delta

print(x,y)
```

```python
plt.plot(x,y,"*")
plt.xlabel("explicative variable 'x' ")
plt.ylabel("target variable 'y'")
plt.title("Scatter plot")
plt.savefig("./intuitive_scatter_bis.png")
```

### La library scikit-learn
 
*La bibliothèque scikit-learn (également appelée sklearn) est l'une des bibliothèques Python les plus populaires pour l'apprentissage automatique (machine learning). Elle fournit des outils efficaces, simples et cohérents pour une variété de tâches d'apprentissage automatique, notamment la classification, la régression, le regroupement, la réduction de dimension, la sélection de modèles et bien d'autres.*

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
#we are using the reshape function to convert the 1D array to a 2D array which is an obligation in scikit-learn LinearRegression() cf official doc
linear_model.fit(x.reshape(-1, 1),y)
```

```python
reg_plot(x,y,linear_model)
plt.savefig("./prediction.png")
```
